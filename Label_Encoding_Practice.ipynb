**Label_Encoding**

*  Converting Categorical Labels into Numeric Values

```python
# Importing the Dependencies
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# Loading the Breast Cancer Dataset into Pandas DataFrames
cancer_data = pd.read_csv("/content/Breast_cancer_dataset.csv")
```

```python
cancer_data.head()
```

```python
# Finding the Counts of Different Labels
cancer_data['diagnosis'].value_counts()
```

```python
# Applying LabelEncoder Function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
# Appending the labels into origional DataFrame
cancer_data['target'] = labels
```

```python
cancer_data.head()
```

```python
# Value counts of (Diagnosis Label)
cancer_data['target'].value_counts()
```

*   0 => Bengin
*   1 => Malignent




**Labeling Encode of Iris Dataset**

```python
# Uploading the Iris Dataset into Pandas DataFrame
iris_data = pd.read_csv('/content/Iris.csv')
```

```python
iris_data.head()
```

```python
# Counting the label [Species]
iris_data['Species'].value_counts()
```

```python
# Applying the LabelEncode Function
label_encode_iris = LabelEncoder()
```

```python
labels = label_encode_iris.fit_transform(iris_data.Species)
```

```python
iris_data['target'] = labels
```

```python
# Count Value for appending labels
iris_data['target'].value_counts()
```

```python
# 0 = > Iris-setosa
# 1 => Iris-versicolor
# 2 => Iris-virginica
```
