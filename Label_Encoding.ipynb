**Label Encoding**

*  Converting all labels into Numeric form


```python
# Importing the Dependenices
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

**Label Encoding of Breast Cancer Dataset**

```python
# Loading the data from csv to Pandas DataFrame
cancer_data = pd.read_csv('/content/Breast_cancer_dataset.csv')
```

```python
# Printing first 5 rows of Dataframe
cancer_data.head()
```

```python
# Finding the count of different labels
cancer_data ['diagnosis'].value_counts()
```

```python
# Load the label encoder function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
# Appending the labels to the dataframe
cancer_data['target'] = labels
```

```python
cancer_data.head()
```

```python
# 0 => Bengin
# 1 => Malignent
```

```python
cancer_data['target'].value_counts()
```

**Label Encoding of Iris Dataset**

```python
# Loading the Iris Data into Pandas DataFrames
iris_data = pd.read_csv('/content/Iris.csv')
```

```python
iris_data.head()
```

```python
iris_data['Species'].value_counts()
```

```python
# Loading the Leblel Encoder Function
label_encoder_iris = LabelEncoder()
```

```python
iris_labels = label_encoder_iris.fit_transform(iris_data.Species)
```

```python
iris_data['target'] = iris_labels
```
